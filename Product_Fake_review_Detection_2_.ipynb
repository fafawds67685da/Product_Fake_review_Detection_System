```python
from nltk.stem import WordNetLemmatizer
Le = WordNetLemmatizer()
```

```python
import re
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
corpus =[]
for i in range(0, len(review)):
  f_review = re.sub('[^a-zA-Z]',' ',str(review['text'][i]))
  f_review = f_review.lower()
  f_review = f_review.split()

  f_review = [Le.lemmatize(word) for word in f_review]
  f_review = ' '.join(f_review)
  corpus.append(f_review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, review['text']) if i<1]
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
nltk.download('punkt_tab')
```

```python
words = []
for sent in corpus:
  sent_token = sent_tokenize(sent)
  #print(sent_token)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
import numpy as n
```

```python
model = gensim.models.Word2Vec(words)
model.wv.index_to_key
```

```python
model.epochs
```

```python
model.wv.similar_by_word('good')
```

```python
model.wv['good'].shape
```

```python
def avg_word2vec(doc):
    # remove out-of-vocabulary words
    #sent = [word for word in doc if word in model.wv.index_to_key]
    #print(sent)

    return n.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
                #or [np.zeros(len(model.wv.index_to_key))], axis=0)
```
